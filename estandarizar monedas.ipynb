```python
import pandas as pd
import numpy as np
import requests
import json
import datetime
import time
```

```python
def monto_uf(moneda, monto, tipo_cambio):
    montoUF = 0;
    if (moneda == 'DOLARES'):
        montoUF = monto*tipo_cambio[tipo_cambio.Sigla == 'DOLARES']['Cambio'].values/tipo_cambio[tipo_cambio.Sigla == 'UF']['Cambio'].values
    elif (moneda == 'US$'):
        montoUF = monto*tipo_cambio[tipo_cambio.Sigla == 'DOLARES']['Cambio'].values/tipo_cambio[tipo_cambio.Sigla == 'UF']['Cambio'].values
    elif (moneda == 'UTM'):
        montoUF = monto*tipo_cambio[tipo_cambio.Sigla == 'UTM']['Cambio'].values/tipo_cambio[tipo_cambio.Sigla == 'UF']['Cambio'].values
    else:
        montoUF = monto
    return(float(montoUF));
```
